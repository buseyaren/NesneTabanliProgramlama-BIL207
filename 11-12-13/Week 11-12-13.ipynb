## Miras, Kalıtım - Inheritance

 Biz, programlama yaparken birden fazla class kullanabiliriz. Yazmış olduğumuz bir class‘a ait özellikleri başka bir class‘ta kullanmak istediğimizde bu özellikleri kalıtım yoluyla almaktayız. Bu işleme miras almak da denilmektedir. Tıpkı bir çocuğun annesine ya da babasına ait birkaç özelliği alması gibi class‘lar da özelliklerini birbirlerinden inheritance yoluyla alabilir. 
 
 Child ve parent olmak üzere 2 farklı class yapısını kullanır.

* Ebeveyn sınıf; değişkenlerini ve fonksiyonlarını kalıtım ile kullandırır. (parent)
* Çocuk sınıf; ebeveyn sınıftan kalıtım ile değişken ve methodaları alarak kullanabilir. (child) - Ebeveynden türetilmiş

Bir sınıfta bulunan aynı kodları tekrar tekrar yazmak yerine ilgili sınıfın aynı kodlarını kullanacak alt sınıflar oluşturulabilir.

Miras almada ebeveyn (parent) ve çocuk (child) olmak üzere iki sınıf yapısından bahsedilir. Child class yapısı parent class yapısının niteliklerini veya metotlarını kullanabilen ve dolayısıyla parent class yapısından türetilmesinden ötürü sub class (alt sınıf) olarak da bilinmektedir.

Sınıfları anlattığımız dersimizde initializier (başlatıcı) kelimesinden türetilen __init__ () fonksiyonunun temel yapıcı fonksiyonu olduğundan bahsetmiştik. __init__ () fonksiyonu bir nesne yazıldığında öncelikli olarak çalıştırılırlar.

Hayvan Sınıfı Örneği



```python
#parent class
class Animal:
    def __init__(self):
        print("Animal is created..")
        
    def toString(self):
        print("Animal")
        
    def walk(self):
        print("Animal is walk")
        
#child class
class Monkey(Animal): #Animal sınıfının çocuğu
    def __init__(self):
        super().__init__()  #use init of parent - Parent classının initialize içeriğini childa aktarıyor
        print("Monkey is created..")
        
    def toString(self):
        print("Monkey")
        
    def climb(self):
        print("Monkey can climb")
        
monkey1 = Monkey()
```

```python
monkey1.toString()
```

```python
monkey1.walk()
```

```python
monkey1.climb()
```

## Abstract Classes (Soyut Sınıflar)

Bir ya da birden fazla Sub Class (Alt Sınıf)’ın inheritance (kalıtım/miras) aldıkları class’a Base Class (taban sınıf) ya da Parent Class (Ana Sınıfı) dediğimizi hatırlayalım. Biz, bazı durumlarda taban sınıfımızı soyut hale getirmeyi tercih ederiz. Çünkü; soyut bir sınıf oluşturduğumuzda bu soyut sınıfı miras alan alt sınıfların her biri soyut sınıfa ait olan metotları kullanmak zorundadır. Yukarıdaki sorumuzun cevabını aslında burada almış bulunuyoruz: Zorunluluk.

Bir taban sınıfı oluşturduğumuzu ve bu sınıf içinde de bazı metotlar yazdığımızı düşünelim. Biz, yazmış olduğumuz metotların taban sınıfı miras alan alt sınıflarda da kullanılmasını istediğimizde soyut sınıflara ihtiyaç duyarız. Buradaki “istek” sözü bizleri yanıltmasın, burada bu metotların kullanılmasının zorunlu olmasından bahsediyoruz. Bir örnek ile daha iyi kavrayabiliriz.

Programlamanın önemli ilkelerinden biri olarak bilinen ve Türkçesiyle “Kendini tekrar etme” anlamına gelen DRY (Don’t Repeat Yourself) prensibi, kod tekrarlarından sakınılması gerektiğini savunur.

Oldukça büyük bir projede malumunuz kod kopyalarının çoğalması hataları da beraberinde getirmektedir. Ayrıca programcının tüm sınıfları ayrıntıları ile hatırlaması da oldukça zordur. Bu nedenle, farklı uygulamalar için ortak bir arabirim tanımlama gereksinimi doğmuştur. Soyut sınıflar ile bu tarz bir arabirim kolayca tanımlanabilmektedir.

Soyut sınıflar, içerisinde soyut metot (abstract method) olarak adlandırılan metotları barındıran sınıflardır. Bu metotlar ise kısaca ismi olup içeriği olmaması durumu olarak tanımlanabilirler. Soyut sınıflar, alt sınıflar (child class) için şablon görevi görürler ve kullanılacak olan ortak fonksiyonları kendi içlerinde tutarlar.

Yani özetle, üst sınıf (parent class), alt sınıflar için gerekli olan metotları bir şablon niteliğinde kendi içinde barındırır.

```python
from abc import ABC, abstractmethod  #abstract base class
class Animal(ABC): #super class

    #Abstract sınıflar için kural 1;
    #Animal sınıfına ait hiçbir obje üretilemez. Bu durumun çözülmesi için built-in module kullanılmalıdır.
    #kural 2 - abstract method decorator kullanılacak.
    @abstractmethod
    def walk(self): pass
    
    @abstractmethod
    def run(self): pass      

an1 = Animal()
```

### Örneklendirme hatası alındı.

```python
"Python'da Soyut Sınıflar (Abstract Classes) - Subclasslar için şablon görevi görür ve kullanılacak
"ortak fonks.ları kendi içerisinde tutarlar"
from abc import ABC, abstractmethod
 
class Hayvan (ABC): # parent class (üst sınıf, ebeveyn sınıf) 
    
    @abstractmethod
    def yurumek(self): # alt sınıfta bu metot kullanılmakt zorunda
        pass
 
    def kosmak(self): #alt sınıfta bu metot kullanılmak zorunda değil
        pass

class Kus(Hayvan): # sub class (alt sınıf, çocuk sınıf)
    
    def __init__(self):
        print("Kuş")
        
    def yurumek(self): 
        print("Yürür")
 
 
kus1 = Kus()
```
