#### Sınıf için __init__ yöntemi oluşturma
Sınıf oluşturma görevimize devam ederken, bir sonraki adım “__init__” adlı çok özel bir yöntemin oluşturulmasını içeriyordu. “Bu __init__ yöntemi nedir ve neden böyle adlandırılıyor?” Diye sorabilirsiniz. Bu kavramı kolayca anlamak için Java'daki ana yönteme benzer init yöntemini düşünün.

Herhangi bir Java kodunu analiz edersek, her sınıfla ilişkili ana yöntem olduğunu yapabiliriz. Aynı şey Python'daki konsepttir. Burada “ana” yöntem olarak adlandırmak yerine “__init__” yöntemi olarak bilinir. Bu kodu sınıfımıza ekleyelim.

### Code For Creating __init__( ) method for the class:
Yukarıdaki koddan da görebileceğiniz gibi burada Python'da __init__ metodumuzu oluşturduk. Burada “def” adlı anahtar kelime olan bir şeyi fark ediyoruz. Bu def anahtar kelimesinin buradaki amacı nedir? Temel olarak def, bir nesne için özellikle hangi tanımlama yöntemini kullandığımız python'da bir anahtar kelimedir.

Ayrıca burada bir şeyi fark ediyorsunuz. Varsayılan olarak, python yazdığımız koddaki bazı girintileri hizalıyor. Python'da “Syntax Engine” olarak bilinen şeye tanık oluyoruz. Bu sözdizimi motoru kendi başına çalışır. Ana amacı, kodumuzun düzgün ve temiz görünmesini sağlamaktır.

Sınıfımız için __init__() metodumuzu tanımladığımız için hala tam olarak tanımlanmadı. Yine de, tanımını tamamlamak için __init__( ) yönteminde belirli argümanların iletilmesi gerekir. Bu argümanları __init__( ) yöntemimize ekleyelim.

Burada, __init__ metodunun 4 tane zorunlu argümanının eksik olduğunu bize bildirmektedir. İşte, __init__ metodunun parantezi içindeki değerler, bizim, nesneleremize atadığımız ve bu class içinden bir nesne türetilecekse bu 4 özelliği alması zorunludur dediğimiz özelliklerimizdir.

**Yapıcılar genellikle bir nesneyi başlatmak için kullanılır. Yapıcıların görevi, sınıfın bir nesnesi oluşturulduğunda sınıfın veri üyelerine ilk değeri vermektir (değerleri atamaktır). Python'da __init__( ) yöntemine yapıcı denir ve bir nesne oluşturulduğunda her zaman çağrılır.**

```python
class Dusman:
    #Her masa kırmızı değildir, nesneler de dolayısıyla aynı özelliklere sahip değildir. Bu sebeple init fonksiyonu kullanacağız.
    def __init__(self, isim, kalan_can, saldiri_gucu, mermi_sayisi):
        self.isim = isim
        self.kalan_can = kalan_can
        self.saldiri_gucu= saldiri_gucu
        self.mermi_sayisi = mermi_sayisi
        
dusman1 = Dusman("Hanifi","4","50","17") #object
dusman2 = Dusman("Enes", "5", "50","18") #object
print("Dusman1 yaziliyor.....")
print("\nName:", dusman1.isim, "Kalan Can:", dusman1.kalan_can, 
      "Saldiri gucu:", dusman1.saldiri_gucu, 
      "Mermi sayisi:", dusman1.mermi_sayisi)

print("Dusman2 yaziliyor.....\n")
print("Name:", dusman2.isim, "Kalan Can:", dusman2.kalan_can, 
      "Saldiri gucu:", dusman2.saldiri_gucu, 
      "Mermi sayisi:", dusman2.mermi_sayisi)
#dusman2.print()
```

```python
class Dusman:
    #Her masa kırmızı değildir, nesneler de dolayısıyla aynı özelliklere sahip değildir. Bu sebeple init fonksiyonu kullanacağız.
    def __init__(self):
        self.isim = "Dusman"
        self.kalan_can = 1500
        self.saldiri_gucu= 30
        self.mermi_sayisi = 40
    
    def print(self):
        print("Yazdiriliyor....")
        print("İsim:", self.isim, "Kalan can:", self.kalan_can, "Saldiri gucu:", self.saldiri_gucu, "Mermi sayisi:", self.mermi_sayisi)

dusman1 = Dusman()
dusman2 = Dusman()
print("Dusman1 yaziliyor.....")
dusman1.print()
print("Dusman2 yaziliyor.....")
dusman2.print()
```
